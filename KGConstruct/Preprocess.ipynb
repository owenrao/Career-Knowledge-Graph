# Filter Uncommon Titles

## Filter by non Nan Skills

```python
all_titles = all_titles[(~all_titles["skills"].isnull())]
```

## Filter by Intersection of skills

```python
all_titles[(all_titles["skills"].isnull())&(~all_titles["skills_y"].isnull())]
```

## Combine Columns

```python
all_titles["description"] = all_titles["JD_x"]+"\n"+all_titles["JD_y"]
```

```python
def x_type(a):
  if type(a) is not float:
    if a.endswith("YEARLY"):
      return "Yearly"
    else:
      return "Hourly"
  return a

def y_type(a):
  if type(a) is not float:
    if a.endswith("/yr"):
      return "Yearly"
    else:
      return "Hourly"
  return a
```

```python
all_titles["salaryType_x"] = all_titles["salary_x"].apply(x_type)
```

```python
all_titles["salaryType_y"] = all_titles["salary_y"].apply(y_type)
```

```python
def estimate_salaryYearly(x):
  if type(x.salaryType_y) is not float:
      return float("".join(re.findall(r'\d+',x.salary_y)))
  elif x.salaryType_x == "Yearly":
    try:
      return float("".join(re.findall(r'\d+',x.salary_x)))
    except:
      print(x.salary_x)
      return np.nan
  else:
    return np.nan

def estimate_salaryHourly(x):
  if x.salaryType_x == "Hourly":
    try:
      return float(".".join(re.findall(r'\d+',x.salary_x)))
    except:
      return np.nan
  else:
    return np.nan
```

```python
all_titles["salaryYearly"] = all_titles.apply(estimate_salaryYearly,axis=1)
```

```python
all_titles["salaryHourly"] = all_titles.apply(estimate_salaryHourly,axis=1)
```

```python
all_titles.columns
```

```python
def name_selection(x):
  if type(x.name_x) != float:
    return x.name_x
  else:
    return x.name_y
```

```python
all_titles["name"] = all_titles.apply(name_selection, axis = 1)
```
